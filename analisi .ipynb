```python
import pandas as pd

df_m=pd.read_csv("../../BW Python/melluso.csv")
```

```python
df_m= df_m.replace("Décolleté", "Dècolletè")
df_m= df_m.replace("Infradita", "Infradito")
```

```python
df_l=pd.read_csv("../../BW Python/lumberjack.csv")
df_l
```

```python
df_l= df_l.replace("Scarpa", "Sneakers")
df_l = df_l.replace("Sneaker", "Sneakers")
df_l = df_l.replace("Green boat shoes", "Scarpe da barca")
```

```python
prezzi_m=df_m.groupby("Modello")["Prezzo"].describe()
```

```python
prezzi_m = prezzi_m.loc[:"mean"].round(2)
```

```python
prezzi_m
```

```python
prezzi_l=df_l.groupby(" Modello")[" Prezzo"].describe()
```

```python
prezzi_l = prezzi_l.loc[:"mean"].round(2)
```

```python
prezzi_l
```

```python
merged_df = pd.merge(prezzi_l, prezzi_m, left_on=' Modello', right_on='Modello')
```

```python
merged_df
```

```python

merged_df.plot(x='', kind='bar')
```
